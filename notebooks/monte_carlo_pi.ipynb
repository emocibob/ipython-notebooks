# Monte Carlo approximation of pi

## Read me

The Monte Carlo methods/simulations use random sampling to calculate a certain value.
More on the topic on [internet's favorite encyclopedia](https://en.wikipedia.org/wiki/Monte_Carlo_method).

Here we are going to show how to approximate pi ($\pi$) using a Monte Carlo simulation.
The implementation is written for Python 3 and prior knowledge of the language is recommended.

I posted the code of the figures for the curious but that code is not important - the math and the simulation are.

## The math (and some plots)

Let's say we have a circle inscribed in a square.
If the circle has a radius $r$, then the sides of the square will have a length of $2r$.
In this case the area of the circle is $r^2\pi$ and the area of our square is $(2r)^2$.
Let me do a quick drawing.

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(6, 6)) # figure size in inches
plt.axis([-1, 1, -1, 1]) # x & y axis limit
ax = fig.add_subplot(111) # add a subplot (1x1 grid, 1st subplot)
circle = plt.Circle((0, 0), radius=1, alpha=0.4) # create a circle
ax.add_patch(circle) # add the circle
plt.show()
```

Now let's see what the ratio of the areas of the circle and square is.

$$\frac{r^2\pi}{4r^2} = \frac{\pi}{4}$$

Imagine the figure above was a square dartboard and we were throwing (rather unskillfully) *a lot* of darts at it.
The ones that would end up on the board would be all over it.
We could assume they are randomly distributed on it.

Knowing that, we see how the number of darts in an area is proportional to that area - bigger area means more darts.
We can use that to approximate the circle/square area ratio using the number of darts in those areas.
If we mark the number of darts landed on the board as $throws$ and the number of darts inside the circle as $hits$ we get the following.

$$\frac{hits}{throws} = \frac{\pi}{4}$$

Obviously, with more darts/points we should get a more accurate ratio.

The next drawing shows just a quarter of our board (and darts) so we can simplify the later code, but of course the circle/square area ratio in the drawing remains the same ($\pi/4$).

```python
import numpy as np
import matplotlib.pyplot as plt

throws = 1000
''' x & y coordinates for our darts/points '''
x = np.random.rand(throws)
y = np.random.rand(throws)

fig = plt.figure(figsize=(7, 7))
plt.axis([0, 1, 0, 1])
ax = fig.add_subplot(111)
arc = plt.Circle((0, 0), radius=1, alpha=0.4)
ax.add_patch(arc)
ax.scatter(x, y, color='r', edgecolor='none') # plot all darts/points
plt.show()
```

We know the number of darts thrown but what about the number of darts inside the circle?
There's an easy way of computing that.

In the example above the circle's radius is 1.
If we calculate the distance of a dart (point) from the circle's center - and its *less or equal* then 1 - then the dart belongs inside the circle.
In case you're wondering, there's no practical difference in using *less or equal* or just *less* for our purposes.

Since we're calculating the distance between two points (a dart and the center), we use the appropriate formula.
All we need to know is the $x$ and $y$ coordinates of these two points.

$$ d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2} $$

Since we know that the coordinates of our center are $(0, 0)$ we can simplify the formula.
The dart's coordinates are $(x, y)$.

$$ d = \sqrt{(0 - x)^2 + (0 - y)^2} = \sqrt{x^2 + y^2} $$

OK, so we have the number of darts on the board and the number of darts inside the circle.
We saw before how

$$ \frac{hits}{throws} = \frac{\pi}{4} $$

so if we multiply the equation by 4, we finally get $\pi$.

$$ 4 \times \frac{hits}{throws} = \pi$$

## The simulation

Now we just need the implement in Python 3 what we learned above.
In the following (and prior) example we use the [NumPy](http://www.numpy.org/) package so we don't have to implement certain functions ourself (shorter code, yay!).

First we create two arrays (`x` and `y`) of 500000 random numbers each.
They will hold the $x$ and $y$ coordinates of our darts/points.
Both random coordinates can have values from 0 to 1, just as the darts in the figure earlier.

Then we calculate the number of darts inside the circle the way shown before.
`x*x + y*y`, `np.sqrt` and `d <= 1` operate on all members of the array - they return a new array with new values for each member.

After we calculate `pi` using the formula we presented earlier we also calculate the error (absolute difference from the real value of $\pi$, i.e. `np.pi`).
In the end we just print the values we just computed.

```python
import numpy as np

''' x & y coordinates for our darts/points '''
throws = 500000
x = np.random.rand(throws)
y = np.random.rand(throws)

''' get number of hits '''
d = np.sqrt(x*x + y*y)
hits = np.sum(d <= 1)

''' calculate pi '''
pi = 4 * (hits / throws)
error = abs(np.pi - pi)

print('Approximate pi: ', str(pi))
print('Error:          ', str(error))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.html.widgets import interact, IntSliderWidget, CheckboxWidget

def monteCarloPi(throws, showFig):
    
    x = np.random.rand(throws)
    y = np.random.rand(throws)
    
    d = np.sqrt(x*x + y*y)
    hits = np.sum(d <= 1)

    pi = 4 * (hits / throws)
    error = abs(np.pi - pi)

    print('Approximate pi: ', str(pi))
    print('Error:          ', str(error))
    
    if showFig:
        fig = plt.figure(figsize=(10, 10))
        plt.axis([0, 1, 0, 1])
        ax = fig.add_subplot(111)
        ax.scatter(x, y, s=6, color='r', edgecolor='none')
        arc = plt.Circle((0, 0), radius=1, alpha=0.4)
        ax.add_patch(arc)
        plt.show()

throwsSlider = IntSliderWidget(min=10, max=50000, step=10, value=5000)
showFigCheck = CheckboxWidget(value=False)
interact(monteCarloPi, throws=throwsSlider, showFig=showFigCheck)
```

## Bonus: Why more points doesn't always mean higher precision

In the next code we run the simulation multiple times and then store the error for each number of throws.
We then plot the different numbers of throws versus their resulting errors in calculating $\pi$ (blue line).
Additionally, we add a fitting curve showing a trend in the data (red dashed line).

Let's take a look at the plot below the code.

Initially the accuracy seems to be rising (lower error with more throws) but then it get's less accurate. You can see the red dashed line slowly rising on the right side of the plot.
Should this be happening?

Ideally more throws would mean a better result, but there are some [practical](https://www.goodreads.com/quotes/66864-in-theory-theory-and-practice-are-the-same-in-practice) issues we need to consider:
1. We stated at the beginning that Monte Carlo simulations use *random* sampling.
If you know some programming you should already be aware that most of the time we are dealing with *pseudorandom* numbers and data (not random).
Don't let sneaky function names like `numpy.random.rand` fool you, they're called like that [just because *random* is faster to type than *pseudorandom*](http://www.thefreedictionary.com/probably).
2. An even bigger issue is the limited precision of [floating point](https://en.wikipedia.org/wiki/Floating_point) numbers which we use to represent decimal values.
Small errors can accumulate when dealing with a great number of operations.
